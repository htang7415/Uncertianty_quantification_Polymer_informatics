```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Tg_Original.csv'
data_Tg = pd.read_csv(file_path)
# Print the number of data points in the dataset
num_data_points = len(data_Tg)
print(num_data_points)

# Plot the distribution of DielectricValue
plt.figure(figsize=(10, 6))
plt.hist(data_Tg['Tg'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of Tg Value')
plt.xlabel('Tg Value')
plt.ylabel('Frequency')
# plt.grid(True)
plt.show()
```
